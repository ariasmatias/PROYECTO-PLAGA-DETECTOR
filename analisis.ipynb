```python
import os
import cv2
import csv

folder = "set_extraido"
imagenes = [f for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Archivo CSV donde se guardarán las coordenadas
csv_file = "rois.csv"

with open(csv_file, mode='w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["imagen", "x", "y", "w", "h"])  # cabecera

    for img_name in imagenes:
        path = os.path.join(folder, img_name)
        img = cv2.imread(path)

        # Seleccionar ROI
        roi = cv2.selectROI("Imagen", img, showCrosshair=True, fromCenter=False)
        x, y, w, h = roi
        print(f"{img_name} -> ROI seleccionada: x={x}, y={y}, w={w}, h={h}")

        # Guardar en CSV
        writer.writerow([img_name, x, y, w, h])

        # Dibujar rectángulo y mostrarlo
        if w > 0 and h > 0:
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.imshow("Imagen con ROI", img)
            cv2.waitKey(500)

cv2.destroyAllWindows()
print(f"✅ Todas las coordenadas se guardaron en {csv_file}")

```
